# From the above Heatmap:


1.   Sales and Profit are Moderately Correlated.
2.   Discount and Profit are Negatively Correlated.
3.   Quantity and Profit are less Moderately Correlated.


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
sns.countplot(data['Sub-Category'])
plt.title('Sub-Category',fontsize=30)
```

```python
data_segment = data.groupby(['Segment'])[['Sales','Discount','Profit']].mean()
data_segment
```

```python
#1. sales 2. Discount 3. Profit
data_segment.plot.pie(subplots= True,
                   autopct= '%1.1f%%',
                   figsize= (18,20),
                   startangle= 90,
                   shadow = True,
                   labels = data_segment.index)
plt.title('Segment analysis of Sale, Dicount, Profit')
```

```python
data_quantity = data.groupby(['Quantity'])[['Sales','Discount','Profit']].mean()
data_quantity.head(10)
```

```python
data_quantity.plot.pie(subplots= True,
                    autopct= '%1.1f%%',
                    figsize=(20,20),
                    pctdistance= 0.69,
                    startangle= 90,
                    shadow= True,
                    labels = data_quantity.index)
plt.title('Quantity-wise analysis of Sales, Discount, Profit')
```

13 Number of Quantity is high for sales and Profit.

```python
data_category = data.groupby(['Category'])[['Sales','Discount','Profit']].mean()
data_category
```

```python
data_category.plot.pie(subplots= True,
                    figsize= (18, 20),
                    autopct= '%1.1f%%',
                    labels = data_category.index)
```
